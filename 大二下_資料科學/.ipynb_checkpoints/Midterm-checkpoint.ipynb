```python
from selenium import webdriver
import urllib.request
from time import sleep
import cv2
f=open('d:/config.txt','r',encoding='utf-8')
s2 = [line.strip() for line in f.readlines()]
print(s2)
print()
f.close()
j=0
for i in s2:
    k=0
    browser = webdriver.Chrome()
    browser.get(i)
    images = browser.find_elements_by_tag_name('img')
    for img in images:
        source = img.get_attribute("src")
        name='D:/question1/'+str(j)+str(k)+'.png'
        sleep(1)
        urllib.request.urlretrieve(source,name)
        img = cv2.imread(name)
        img_small = cv2.resize(img,(100,100))
        smallname='D:/question1small/'+str(j)+str(k)+'.png'
        cv2.imwrite(smallname,img_small)
        k+=1
    browser.quit()
    j+=1
    
    
```

```python
from time import sleep
from selenium import webdriver
from bs4 import BeautifulSoup
browser = webdriver.Chrome()
browser.get('https://tip.railway.gov.tw/tra-tip-web/tip?lang=EN_US')
browser.find_element_by_id('startStation').send_keys('1040-Shulin')
browser.find_element_by_id('endStation').send_keys('7000-Hualien')
sleep(3)
browser.find_element_by_id('startTime').send_keys('06:00')
browser.find_element_by_id('endTime').send_keys('18:00')
sleep(3)
browser.find_element_by_xpath("//input[@value='Inquiry']").click()
sleep(3)
result_html=browser.find_element_by_css_selector("*").get_attribute("outerHTML")
bs=BeautifulSoup(result_html,'lxml')
bs
```

```python
x=bs.find({'table'},{'class':'itinerary-controls'})
x
```

```python
import pandas as pd
bus_table = pd.read_html(result_html)
print(bus_table)
```

```python
bus_df = bus_table[0]
bus_df
```

```python
df=bus_df[bus_df.index%5==0]
df
```

```python
df=df.drop('Via',axis = 1)
df=df.drop('Detailedinformation',axis = 1)
df=df.drop('Order Tickets',axis = 1)
df=df.drop('Unnamed: 9',axis = 1)
df=df.drop('Unnamed: 10',axis = 1)
df
```

```python
df=df.reset_index(drop=True)
df
```

```python
df.to_csv("d:/midterm.csv")
```
