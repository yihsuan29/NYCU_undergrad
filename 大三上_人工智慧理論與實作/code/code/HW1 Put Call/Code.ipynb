```python
import numpy as np
import matplotlib.pyplot as plt
def LongCall(ST,K,call):
    return np.maximum(ST-K,0)-call
def ShortCall(ST,K,call):
    return -(np.maximum(ST-K,0)-call)
def LongPut(ST,K,put):
    return np.maximum(K-ST,0)-put
def ShortPut(ST,K,put):
    return -(np.maximum(K-ST,0)-put)
ST=np.arange(12000,16001)
```

```python
# Q1 short call
SC13500=ShortCall(ST,13500,345)
SC14000=ShortCall(ST,14000,120)
plt.figure(figsize=(9,9))
plt.plot(ST,SC13500,'b-',ST,SC14000,'c-',[12000,16001],[0,0],'k--',[13500,13500],[2000,-2500],'k--')
```

```python
# Q1 long put
LP14300=LongPut(ST,14300,800)
LP15000=LongPut(ST,15000,1470)
plt.figure(figsize=(9,9))
plt.plot(ST,LP14300,'r-',ST,LP15000,'m-',[12000,16001],[0,0],'k--',[13500,13500],[2000,-2500],'k--')
```

```python
#bear spread
SC13500=ShortCall(ST,13500,345)
LC14000=LongCall(ST,14000,120)
LC14500=LongCall(ST,14500,28)
plt.figure(figsize=(9,9))
plt.plot(ST,SC13500+LC14000,'g-',ST,SC13500+LC14500,'y-',[12000,16001],[0,0],'k--',[13500,13500],[2000,-2500],'k--')
```

```python
plt.figure(figsize=(9,9))
plt.plot(ST,SC13500,'b-',ST,SC14000,'c-',ST,LP14300,'r-',ST,LP15000,'m-',ST,SC13500+LC14000,'g-',ST,SC13500+LC14500,'y-',[12000,16001],[0,0],'k--',[13500,13500],[2000,-2500],'k--')
```

```python
#Q2 butterfly
LP13700=LongPut(ST,13700,388)
SP14000=ShortPut(ST,14000,565)
LP14300=LongPut(ST,14300,800)

LP13400=LongPut(ST,13400,255)
LP14600=LongPut(ST,14600,1090)
plt.figure(figsize=(9,9))
plt.plot(ST,LP13700+2*SP14000+LP14300,'b-',ST,LP13400+2*SP14000+LP14600,'c-',[12000,16001],[0,0],'k--',[14000,14000],[750,-1250],'k--')
```

```python
#Q2 straddle
SC14000=ShortCall(ST,14000,120)
SP14000=ShortPut(ST,14000,565)
plt.figure(figsize=(9,9))
plt.plot(ST,SC14000+SP14000,'r-',[12000,16001],[0,0],'k--',[14000,14000],[750,-1250],'k--')
```

```python
#Q2 strangle
SC14200=ShortCall(ST,14200,69)
SP13800=ShortPut(ST,13800,439)
SC14500=ShortCall(ST,14500,28)
SP13500=ShortPut(ST,13500,291)
plt.figure(figsize=(9,9))
plt.plot(ST,SC14200+SP13800,'g-',ST,SC14500+SP13500,'y-',[12000,16001],[0,0],'k--',[14000,14000],[750,-1250],'k--')
```

```python
#Q3 arbitrage
call_price=[675,610,525,448,395,345,289,238,194,153,120,92,69,52,39.5,28,20.5,16,12,9.8,8.5]
put_price=[139,163,189,221,255,291,338,388,439,505,565,640,730,800,900,995,1090,1180,1290,1380,1470]
PF=[]
for i in range(21):
    x=13000+i*100
    tmpSC=ShortCall(13000,x,call_price[i])
    tmpLP=LongPut(13000,x,put_price[i])
    PF.append(tmpSC+tmpLP)
print(PF.index(min(PF)))
print(PF.index(max(PF)))
mini=13000+PF.index(min(PF))*100
maxi=13000+PF.index(max(PF))*100
#
print("LC:"+str(mini))
print("SP:"+str(mini))
#
print("SC:"+str(maxi))
print("LP:"+str(maxi))
print("profit:"+str(PF[PF.index(max(PF))]-PF[PF.index(min(PF))]))
```
