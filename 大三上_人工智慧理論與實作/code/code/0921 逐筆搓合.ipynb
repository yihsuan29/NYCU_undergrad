```python
from types import BuiltinMethodType
T=int(input())
for t in range(T):
    N=int(input())
    buy=[]
    sell=[]
    stockprice=-1
    for n in range(N):
        strlist=input().split()
        share=int(strlist[1])  #量
        price=int(strlist[-1]) #價
        if(strlist[0]=="buy"):
            while len(sell)>0:
                order = sell[0]
                if order[0]>price:
                    break
                delamount=min(share,order[1])
                stockprice=order[0]
                order[1]-=delamount
                share-=delamount
                #單筆賣單結束
                if order[1]==0:
                    del sell[0]
                #買賣搓合完畢
                if share==0:
                    break
            if share>0:
                i=0
                #到尾or比較便宜
                while i<len(buy) and price<=buy[i][0]:
                    i+=1
                #同價加單
                if(i<len(buy) and price==buy[i][0]):
                    buy[i][1]+=share
                #新價格
                else:
                    buy.insert(i,[price,share])

        else:
            while len(buy)>0:
                order = buy[0]
                if order[0]<price:
                    break
                delamount=min(share,order[1])
                stockprice=price
                order[1]-=delamount
                share-=delamount
                #單筆賣單結束
                if order[1]==0:
                    del buy[0]
                #買賣搓合完畢
                if share==0:
                    break
            if share>0:
                i=0
                #到尾or比較便宜
                while i<len(sell) and price>=sell[i][0]:
                    i+=1
                #同價加單
                if(i<len(sell) and price==sell[i][0]):
                    sell[i][1]+=share
                #新價格
                else:
                    sell.insert(i,[price,share])
        if len(sell)==0:
            print('-',end=' ')
        else:
            print(sell[0][0],end=' ')
        if len(buy)==0:
            print('-',end=' ')
        else:
            print(buy[0][0],end=' ')
        if stockprice==-1:
            print('-')
        else:
            print(stockprice)
```
