```python
import numpy as np
import matplotlib.pyplot as plt
def LongCall(ST,K,call):
    return np.maximum(ST-K,0)-call
def ShortCall(ST,K,call):
    return -(np.maximum(ST-K,0)-call)
def LongPut(ST,K,put):
    return np.maximum(K-ST,0)-put
def ShortPut(ST,K,put):
    return -(np.maximum(K-ST,0)-put)
ST=np.arange(15001,18001)
LC16200=LongCall(ST,16200,372)
plt.plot(ST,LC16200,'b-',[15001,18000],[0,0],'k--')
```

```python
SP16400=ShortPut(ST,16400,339)
SP16300=ShortPut(ST,16300,299)
plt.plot(ST,SP16400,'b-',ST,SP16300,'r-',[15001,18000],[0,0],'k--')
```

```python
#bull spread
SC16500=ShortCall(ST,16500,199)
LC16200=LongCall(ST,16200,372)
plt.plot(ST,SC16500,'g-',ST,LC16200,'b-',ST,SC16500+LC16200,'r-',[15001,18000],[0,0],'k--')
```

```python
#butterfly
LP16200=LongPut(ST,16200,265)
SP16400=ShortPut(ST,16400,339)
LP16600=LongPut(ST,16600,463)
plt.plot(ST,LP16200+2*SP16400+LP16600,'b-',[15001,18000],[0,0],'k--')
```

```python
#straddle
LC16400=LongCall(ST,16400,248)
LP16400=LongPut(ST,16400,339)
plt.plot(ST,LC16400+LP16400,'b-',[15001,18000],[0,0],'k--')
```
