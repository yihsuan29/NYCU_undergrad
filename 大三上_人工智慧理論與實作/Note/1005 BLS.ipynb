```python
#black-scholes
import math
from scipy.stats import norm

def blscall(S,K,T,r,vol):
    d1=(math.log(S/K)+(r+vol**2/2)*T)/(vol*T**(1/2))
    d2=d1-vol*T**(1/2)
    c=S*norm.cdf(d1)-K*math.exp(-r*T)*norm.cdf(d2)
    return c
S=13600
K=13000
T=0.08
r=0.01
vol=0.2
blscall(S,K,T,r,vol)
```

```python
for n in range(0,200):
    vol=0.148+n*0.00001
    print(vol,blscall(S,K,T,r,vol))
```

```python
def BisectionBLS(S,K,T,r,call):
    left=1e-8
    right=1
    threshold=1e-5
    while right-left>=threshold:
        middle=(left+right)/2
        if((blscall(S,K,T,r,left)-call)*(blscall(S,K,T,r,middle)-call)<0):
            right=middle
        else:
            left=middle
    return (left+right)/2

BisectionBLS(S,K,T,r,650)  
```
